## Exploratory Data Analysis (EDA)

### BMI: 
BMI stands for Body Mass Index, a calculation that estimates body fat based on height and weight. BMI is a screening tool that helps determine if someone's weight may put them at risk for health problems.

**Calculation:** BMI is calculated by dividing weight in kilograms by height in meters squared (BMI = kg/m²). 



```python
### New BMI column

# Convert height from centimeters to meters
df['height_m'] = df['HE_ht'] / 100

# Calculate BMI and create a new column
df['BMI'] = df['HE_wt'] / (df['height_m'] ** 2)

# Show the new BMI values
df['BMI'].head()
```

## Hypothesis Testing

### Propose Hypothesis

**Hypothesis 1:** Participants who walk more than 3 days per week are less likely to have prediabetes.

**Hypothesis 2:** Unemployed participants are more likely to have prediabetes than employed participants.

**Hypothesis 3:** Participants with prediabetes have a higher average BMI than those without.

**Hypothesis 4:** Participants who do strenght training at least 2 times per week have a lower risk of diabetes.

### Analysis

#### **Hypothesis 1:**

Participants who walk more than 3 days per week are less likely to have prediabetes.

**Variables:**
* BE3_31: Number of days walking per week
* diabetes: Diabetes status of the participant

**Hypothesis:**
* H0: There is no difference in diabetes rates between those who walk more than 3 days and those who don't
* H1: Participants who walk more than 3 days are less likely to have prediabetes

```python
import scipy.stats as stats
from scipy.stats import chi2_contingency

df['walk_3plus'] = df['BE3_31'] > 3
contingency1 = pd.crosstab(df['walk_3plus'], df['diabetes'])

chi2_1, p_1, _, _ = chi2_contingency(contingency1)
print("Chi-square p-value:", p_1)
```

**Interpretation:** Because p-value < 0.05 we can reject the null hypothesis. It means that there is a statistically significant association between walking more than 3 days per week and a lower chance to have prediabetes.

#### **Hypothesis 2:**

Unemployed participants are more likely to have prediabetes than employed participants.

**Variables:**
* EC1_1: Employment status of the participant
* diabetes: Diabetes status of the participant

**Hypothesis:**
* H0: Employment status is not associated with prediabetes
* H1: Unemployed participants are more likely to have prediabetes

```python
contingency2 = pd.crosstab(df['EC1_1'], df['diabetes'])

chi2_2, p_2, _, _ = chi2_contingency(contingency2)
print("Chi-square p-value:", p_2)
```

**Interpretation:** Because p-value > 0.05 we can't reject the null hypothesis. It means that there is not significant relationship between employment and prediabetes.

#### **Hypothesis 3:**

Participants with prediabetes have a higher average BMI than those without.

**Variables:**
* BMI
* diabetes: Diabetes status of the participant

**Hypothesis:**
* H0: There is no difference in average BMI between normal and prediabetic participants
* H1: The average BMI is higher among prediabetic participants

```python
from scipy.stats import ttest_ind

# Separate the BMI data into two groups
bmi_normal = df[df['diabetes'] == 0]['BMI']
bmi_prediabetes = df[df['diabetes'] == 1]['BMI']

# t-test
t_stat, p_value = ttest_ind(bmi_prediabetes, bmi_normal, equal_var=False)

print("T-test p-value:", p_value)
```

**Interpretation:** Because p-value < 0.05 we can reject the null hypothesis. It means that there is a statistically significant difference in BMI, and participants with prediabetes have a higher average BMI

#### **Hypothesis 4:**

Participants who do strenght training at least 2 times per week have a lower risk of diabetes.

**Variables:**
* BE5_1: Number of days strength training per week
* diabetes: Diabetes status of the participant

**Hypothesis:**
* H0: Strength training frequency is not related to prediabetes
* H1: Doing strength training more than 2 times per week is associated with lower prediabetes risk

```python
# Create a binary variable: strength training ≥ 2 times per week
df['strength_2plus'] = df['BE5_1'] >= 2

# Contingency table
contingency4 = pd.crosstab(df['strength_2plus'], df['diabetes'])

# Chi-square test
chi2_4, p_4, _, _ = chi2_contingency(contingency4)

print("Chi-square p-value:", p_4)
```

**Interpretation:** Because p-value < 0.05 we can reject the null hypothesis. It means that there is a statistically significant evidence that participants who train at least 2 times per week are less likely to have prediabetes

## Correlation Testing

#### **1. Pearson correlation:** Continuous vs Continuous

Is there a correlation between age and BMI?

**Hypothesis:**
* Ho: There is no linear correlation between age and BMI
* H1: There is a linear correlation between age and BMI

```python
from scipy import stats

age_corr, age_pvalue = stats.pearsonr(df['age'], df['BMI'])
print(f"Pearson Correlation Coefficient: {age_corr}")
print(f"P-Value: {age_pvalue}")
```

**Visualization**

```python
sns.regplot(x='age', y='BMI', data=df, line_kws={"color": "orange"})
plt.title("Age vs. BMI")
plt.xlabel("Age")
plt.ylabel("BMI")
plt.show()
```

#### Interpretation:

**Correlation:** The Pearson correlation coefficient (r = 0.11) indicates a very weak positive linear relationship between age and BMI. This suggests that as age increases, BMI tends to increase as well, but the relationship is not strong.

**P-value:** Since the p-value < 0.05, we reject the null hypothesis, indicating that the relationship between age and BMI is statistically significant, even though it is weak.

#### **2. Spearman correlation:** Rank-based (Monotonic)

Is there a monotonic relationship between walking frequency and moderate-intensity leisure activity?

**Hypothesis:**
* Ho: No monotonic relationship between BE3_31 and BE3_86
* H1: A monotonic (increasing or decreasing) relationship exists

```python
walk_corr, walk_pvalue = stats.spearmanr(df['BE3_31'], df['BE3_86'])
print(f"Spearman Correlation Coefficient: {walk_corr}")
print(f"P-Value: {walk_pvalue}")
```

**Visualization**

```python
sns.regplot(x='BE3_31', y='BE3_86', data=df, line_kws={"color": "red"})
plt.title("Walking Days vs. Moderate-Intensity Leisure Activity")
plt.xlabel("Walking Days per Week")
plt.ylabel("Moderate Leisure Activity Days")
plt.show()
```

#### Interpretation:

**Correlation:** The Spearman correlation coefficient (ρ = 0.19) indicates a weak positive monotonic relationship between walking frequency and moderate intensity leisure activity. This means that participants who walk more often also tend to engage more in moderate intensity leisure activities, but the relationship is not strong.

**P-value:** Since the p-value < 0.05, we reject the null hypothesis, indicating that the relationship is statistically significant, even though it is weak.

#### **3. Point-Biserial correlation:** Binary vs Continuous

Is being prediabetic related to strength training days per week?

**Hypothesis:**
* Ho: There is no correlation between diabetes status and strength training frequency
* H1: There is a correlation between diabetes status and strength training frequency

```python
from scipy.stats import pointbiserialr

pb_corr, pb_pvalue = pointbiserialr(df['diabetes'], df['BE5_1'])
print(f"Point-Biserial Correlation Coefficient: {pb_corr}")
print(f"P-Value: {pb_pvalue}")
```

#### Interpretation:

**Correlation:** The Point-Biserial correlation coefficient (r = -0.03) indicates a very weak negative relationship between being prediabetic and the number of days participants perform strength training. In other words, those with prediabetes tend to engage less in strength training, but the effect is minimal.

**P-value:** Since the p-value < 0.05, we reject the null hypothesis, indicating that the relationship is statistically significant, even though it is weak.

#### **4. Phi correlation:** Binary vs Binary

Is there a relationship between employment status and weight control through exercise?

**Hypothesis:**
* Ho: There is no association between employment and exercising for weight control
* H1: There is an association

```python
import numpy as np
from scipy.stats import chi2_contingency

phi_table = pd.crosstab(df['EC1_1'], df['BO3_01'])
chi2_phi, p_value_phi, _, _ = chi2_contingency(phi_table)
phi_coeff = np.sqrt(chi2_phi / df.shape[0])

print(f"Phi Coefficient: {phi_coeff}")
print(f"Chi-square P-Value: {chi2_phi}")
```

#### Interpretation:

**Correlation:** The Phi coefficient (φ = 0.0) indicates no association between employment status and using exercise as a method for weight control. The two variables appear to be statistically independent.

**P-value:** Despite the p-value being 0.0, which normally suggests a statistically significant result, the Phi coefficient of 0 tells us that there is no meaningful relationship between the two variables, so the result is not practically significant.

#### **5.  Kendall’s Tau:** Monotonic (non-parametric)

Is there a monotonic relationship between high-intensity leisure activity and strength training?

**Hypothesis:**
* Ho: No monotonic relationship exists
* H1: A monotonic relationship exists

```python
tau_corr, tau_pvalue = stats.kendalltau(df['BE3_76'], df['BE5_1'])
print(f"Kendall’s Tau Correlation Coefficient: {tau_corr}")
print(f"P-Value: {tau_pvalue}")
```

**Visualization**

```python
sns.regplot(x='BE3_76', y='BE5_1', data=df, line_kws={"color": "blue"})
plt.title("High-Intensity Leisure Activity vs. Strength Training")
plt.xlabel("High-Intensity Leisure Days")
plt.ylabel("Strength Training Days")
plt.show()
```

#### Interpretation:

**Correlation:** The Kendall’s Tau coefficient (τ = 0.29) indicates a moderate positive monotonic relationship between the number of days participants engage in high-intensity leisure activity and the number of days they perform strength training. This suggests that those who frequently engage in high-intensity leisure activities also tend to do more strength training.

**P-value:**  Since the p-value < 0.05, we reject the null hypothesis, indicating that the relationship is statistically significant.

## Interpretation and Real-World Application

#### Interpretation:

* BMI and Age: Although the Pearson correlation between age and BMI was statistically significant, the relationship was very weak (r ≈ 0.11), suggesting that while BMI tends to slightly increase with age, other factors likely play a more substantial role.

* Prediabetes and Strength Training: The Point-Biserial correlation indicated a very weak but statistically significant negative relationship between strength training frequency and prediabetes status. This suggests that even minimal increases in physical activity might offer protective effects against prediabetes.

* Employment and Health Behavior: The Phi coefficient showed no meaningful association between employment status and whether participants exercised for weight control, despite a statistically significant Chi-square result. This highlights the importance of not only focusing on p-values and considering effect size.

* Data-Informed Health Campaigns: Weak or moderate correlations across variables like walking, leisure activity, and strength training indicate that integrated exercise plans and combining different types of activity, could yield better health results.

This analysis suggests that simple, measurable lifestyle behaviors such as walking, strength training, and engaging in leisure activity are modestly but meaningfully associated with better health outcomes.
